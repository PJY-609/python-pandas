## How to select or remove certain rows and columns using pandas python
* **```df.drop(['attr1', 'attr2',...], axis= , inplace=True)```**
* **```df.iloc[:,i:j]```**  
* **```df.iloc[lamda x: x.index...]```**  
  *pay attention to x.index*

````python
import pandas as pd
df = pd.read_csv('D:/ufo.csv')
df.head()
````

````python
df.drop('Colors Reported', axis=1, inplace=True)
df.head()
````

````python
df.drop(['City', 'State'], axis=1, inplace=True)
df.head()
````

````python
df = pd.read_csv('D:/ufo.csv')
df.head()
````

````python
df.iloc[0:4].head()
````

````python
df.iloc[[0, 4]].head()
````

````python
df.iloc[[True, False, True, False, True]].head()
````

````python
# pay attention to x.index
df.iloc[lambda x: x.index % 2 == 0].head()
````

````python
df = df.iloc[:, 1:4]
df.head()
````
