````python
import pandas as pd
df = pd.read_csv('D:/imdb_1000.csv')
df.head()
````

````python
df[df.actors_list.str.contains('Tom Hanks', regex=True)].head()
````

````python
df[df.actors_list.str.contains('matt damon', case=False)].head()
````

````python
df[df.actors_list.str.contains('Tom Hanks|Al Pacino')].head()
````

````python
word_list = ['Tom Hanks', 'Al Pacino']
````

````python
import re
pat = '|'.join(map(re.escape, word_list))
df[df.actors_list.str.contains(pat, case=False)].head()
````

````python
pat = r'Tom Hanks.*?Tim Allen'
p = re.compile(pat, flags=re.IGNORECASE)
df[[bool(p.search(x)) for x in df['actors_list']]]
````

````python
import ast
name_set = {'Tom Hanks', 'Tim Allen'}
````

````python
df[[name_set.issubset(set(ast.literal_eval(x))) for x in df['actors_list']]].head()
````

````python
df[(df.actors_list.str.contains('Tom Hanks'))&(df.actors_list.str.contains('Tim Allen'))]
````
