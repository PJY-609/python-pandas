## How to filter the rows using pandas python
* **```boolean = df.attr.apply(lambda x: True if x... else False)```**  
    **```df[boolean]```**
* **```df[df.attr >/</==/!= ...]```**
* **```df.loc[df.attr >/</==/!= ..., 'attr_i']```**
* **```df.attr.str.contains('string')```**

````python
import pandas as pd
df = pd.read_csv('D:/imdb_1000.csv')
df.head()
````

````python
is_long = df.duration.apply(lambda x: True if x>200 else False)
df[is_long].head()
````

````python
df[df.duration>200].head()
````

````python
df.loc[df.duration>200, 'title']
````

````python
genre = df.genre.apply(lambda x: True if x in ['Adventure','Drama'] else False)
df[genre].head()
````

````python
th = df.actors_list.str.contains('Tom Hanks',regex=True)
df[th].head()
````
